```python
import scvelo as scv
scv.set_figure_params()
```

```python
mode = 'SCT'

param = 'pc10_res0.1'

file = 'HCC.final.'+mode+'_'+param

dir = '../../../../HCC_Stem_scRNAseq'
data_dir = dir+'/data/Figure_6_and_S6/6E_S6D_scVelo_RNA_velocity/processed_data_for_scVelo'

types = ['Neg', 'Pos']
#types = ['Neg']

mode = 'stochastic'
reduct = 'umap_cell_embeddings'
group = 'seurat_clusters'
```

```python
for type in types:
    clustering_file = data_dir+'/'+file+'.'+type+'.loom'
    adata = scv.read(clustering_file,cache=True)
    
    loom_file = data_dir+'/'+type+'.loom'
    ldata = scv.read(loom_file, cache=True)

    adata = scv.utils.merge(adata, ldata)
    
    scv.pp.filter_and_normalize(adata, min_shared_counts=20, n_top_genes=2000)
    
    scv.pp.moments(adata, n_pcs=30, n_neighbors=30)
    
    scv.tl.velocity(adata, mode=mode)
                                
    scv.tl.velocity_graph(adata)
                                
    scv.pl.velocity_embedding_stream(adata, basis=reduct, color=group, size=100, title=type, fontsize=14, linewidth=1)
```
